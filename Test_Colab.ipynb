The next cell creates a random chart.

````python
import numpy as np
from matplotlib import pyplot as plt

ys = 40 + np.random.randn(100)
x = [x for x in range(len(ys))]
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Random Example Chart")
plt.show()
````
